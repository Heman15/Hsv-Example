In this code we will learn how hsv works

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2                            #importing liberary 
%matplotlib inline
```

```python
img =  cv2.imread("C:\\Users\\saini\\Desktop\\imagesc\\1\\img26.jpg") #Reading image
print(img.shape)
```

```python
plt.imshow(img)
```

```python
img_copy = np.copy(img)
img_copy  = cv2.cvtColor(img_copy , cv2.COLOR_BGR2RGB)        #converting bgr to RGB
plt.imshow(img_copy)
```

```python
r = img_copy[:,:,0]
g = img_copy[:,:,1]
b = img_copy[:,:,2]
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (20,10))
                                                 
ax1.set_title ('Red')
ax1.imshow(r,cmap = "gray")

ax2.set_title ('Green')
ax2.imshow(g,cmap = "gray")
                                                        #PLOTING images for each color channels or we can 
                                                        #isolating each color channels
ax3.set_title ('Blue')
ax3.imshow(b,cmap = "gray")
```

# HSV 

```python
hsv= np.copy(img)
hsv = cv2.cvtColor(hsv, cv2.COLOR_RGB2HSV)
plt.imshow(hsv)                         #IN this we are convertingRGB TO HSV
```

```python
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]
fig,(ax1,ax2,ax3 )= plt.subplots(1,3,figsize = (20,10)) 
ax1.set_title ('hue')
ax1.imshow(h,cmap = "gray")

ax2.set_title ('saturation')
ax2.imshow(s,cmap = "gray")

ax3.set_title ('value')
ax3.imshow(v,cmap = "gray")

#Ploting each value 
```

```python
lower_green = np.array([100, 100, 0])
upper_green = np.array([255, 230, 255])            #setting upper and lower limit for green color 
```

```python
mask_rgb = cv2.inRange(img_copy,lower_green,upper_green)
mask_image = np.copy(img_copy)
plt.imshow(mask_rgb)
```

```python
mask_image[mask_rgb== 0] = [0,0,0]
```

```python
plt.imshow(mask_image)      #it will rgb shows mix colors
```
